```python
# -*- coding: utf-8 -*-
"""
@author: Anonym
"""
import pandas as pd 
import matplotlib.pyplot as plt
import scipy.stats as stats
import scikit_posthocs as sp

filename = "https://docs.google.com/spreadsheets/d/e/2PACX-1vTl_-1qTINIeQGcoj7HljbmXXrmp44DCLfjrVv6_cn0aqYK9Qc9LDGFGPRN2NKGpofBsamXnMadgrLh/pub?output=csv"

data = pd.read_csv(filename)

variables = ["Score", "Task_time" ,"Nb_click"]

s2DU_data = data[data.Surface.str.contains("2D_uniform")]
s2DC_data = data[data.Surface.str.contains("2D_central")]
s2DP_data = data[data.Surface.str.contains("2D_peripheral")]
sHalfSphere_data = data[data.Surface.str.contains("HalfSphere")]
sCone_data = data[data.Surface.str.contains("Cone")]
sTorus_data = data[data.Surface.str.contains("Torus")]

"""
Vérification de la première condition ANOVA :
les distributions ne suivent une loi normale que si p > .05
"""
def RunShapiro(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable):
    print("\n**** Running Shapiro-Wilk test ****\n")
    rep = True
    f, p = stats.shapiro(s2DU[variable])
    print("#2D uniform --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    if p < .05:
        rep = False
    f, p = stats.shapiro(s2DC[variable])
    print("#2D central --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    if p < .05:
        rep = False
    f, p = stats.shapiro(s2DP[variable])
    print("#2D peripheral --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    if p < .05:
        rep = False
    f, p = stats.shapiro(sCone[variable])
    print("#Cone --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    if p < .05:
        rep = False
    f, p = stats.shapiro(sTorus[variable])
    print("#Torus --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    if p < .05:
        rep = False
    f, p = stats.shapiro(sHalfSphere[variable])
    print("#Halfsphere --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    if p < .05:
        rep = False
    
    return rep


"""
une différence est statistiquement significative si p < .05
"""
def RunKruskal(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable):
    print("\n**** Running Kruskal test ****\n")
    f, p = stats.kruskal(s2DU[variable], s2DC[variable], s2DP[variable], sCone[variable], sTorus[variable], sHalfSphere[variable])
    print("Kruskal --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    
    return (f, p)

"""
une différence est statistiquement significative si p < .05
"""
def RunANOVA(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable):
    print("\n**** Running Oneway ANOVA test ****\n")
    f, p = stats.f_oneway(s2DU[variable], s2DC[variable], s2DP[variable], sCone[variable], sTorus[variable], sHalfSphere[variable])
    print("Oneway ANOVA --> F statistic = {:5.6f} and probability p = {:5.6f}\n".format(f, p))
    
    return (f, p)
    

def RunStats(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable):
    f, p = .0, .0
    r = RunShapiro(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable)
    if r == True:
        f, p = RunANOVA(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable) 
    else:
        f, p = RunKruskal(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable)
        
    if p < .05:
        print("There is a difference that is statistically significant\n")
    else:
        print("The difference is not statistically significant")
        
    return (f, p)

"""
Pairwise comparison
"""
def PostHoc(taskId, variable):
    print("\n**** Pairwise comparison ****\n")
    x_data = data[data.Task_id == taskId]
    data_mult = x_data[["Surface", variable]]
    ph = sp.posthoc_dunn(data_mult, val_col = variable, group_col = "Surface", p_adjust = "holm")
    # Format: diagonal, non-significant, p<0.001, p<0.01, p<0.05
    #cmap = ['1', '#fb6a4a',  '#08306b',  '#4292c6', '#c6dbef']
    #heatmap_args = {'cmap': cmap, 'linewidths': 0.25, 'linecolor': '0.5', 'clip_on': False, 'square': True, 'cbar_ax_bbox': [0.80, 0.35, 0.04, 0.3]}
    #sp.sign_plot(ph, **heatmap_args)    
    print(ph)


"""
Box plot
"""
def BoxPlot(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, variable):
    print("\n**** Box plot ****\n")
    fig, ax = plt.subplots()
    fig.set_figheight(5)
    fig.set_figwidth(8)
    ax.set_title("%s : Descriptive representation (mean in red dashes, median in purple)"%variable)
    bp = ax.boxplot((s2DU[variable], s2DC[variable], s2DP[variable], sCone[variable], sTorus[variable], sHalfSphere[variable]), vert=False, showmeans=True, meanline=True,
               labels=("2D Uniform", "2D Central", "2D Peripheral", "Cone", "Torus", "Halfsphere"), patch_artist=True,
               medianprops={'linewidth': 2, 'color': 'purple'},
               meanprops={'linewidth': 2, 'color': 'red'})
    #plt.grid(True)
    for box in bp['boxes']:
        box.set(facecolor='lightblue')
    plt.yticks(fontsize=14)
    plt.show()
    

"""
the main function
"""
def main():
    task = [1, 2, 3]
    for t in task:
        print("\n**********\n \tRunning stats for variables in task %i\n**********"%t)
        try:
            s2DU = s2DU_data[s2DU_data.Task_id == t]
            s2DC = s2DC_data[s2DC_data.Task_id == t]
            s2DP = s2DP_data[s2DP_data.Task_id == t]
            sHalfSphere = sHalfSphere_data[sHalfSphere_data.Task_id == t]
            sCone = sCone_data[sCone_data.Task_id == t]
            sTorus = sTorus_data[sTorus_data.Task_id == t]
            
            for v in variables:
                print("\n**********\n Running stats for %s\n**********"%v)
                f, p = RunStats(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, v)
                if p < .05:
                    PostHoc(t, v) # pairwise comparison
                BoxPlot(s2DU, s2DC, s2DP, sHalfSphere, sCone, sTorus, v) # box plot
        except Exception as e:
            raise e
        


if __name__ == '__main__':
    main()   
```
